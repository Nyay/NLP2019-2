Создадим функцию нормализации текста для нашей задачи

```python
def remove_tags(text):
    return re.sub(r'<[^>]+>', '', text)

def normalize(text):
    
    words = [word.strip(punct) for word in remove_tags(text.lower()).split()]
    words = [morph.parse(word)[0] for word in words if word and word not in get_stop_words('russian')]
    words = [stemmer.stem(word.normal_form) for word in words if word.tag.POS == 'NOUN']
            
    return words
```

## Время для NMF

```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
```

```python
stexts = [' '.join(text) for text in ngrammed_texts]
```

```python
vectorizer = TfidfVectorizer(max_features=30000, min_df=4, max_df=0.3, lowercase=True)
X = vectorizer.fit_transform(stexts)
```

______

```python
model = NMF(n_components=60)
model.fit(X)
```

```python
def get_nmf_topics(model, n_top_words):
    
    #id слов.
    feat_names = vectorizer.get_feature_names()
    
    word_dict = {};
    for i in range(30):
        
        #топ n слов для темы.
        words_ids = model.components_[i].argsort()[:-n_top_words - 1:-1]
        words = [feat_names[key] for key in words_ids]
        word_dict['Topic # ' + '{:02d}'.format(i+1)] = words;
    
    return pd.DataFrame(word_dict);
```

```python
get_nmf_topics(model, 10)
```

```python
model2 = NMF(n_components=60, init='nndsvd')
model2.fit(X)
```

```python
model3 = NMF(n_components=60, init='nndsvdar', solver='mu')
model3.fit(X)
```

```python
get_nmf_topics(model2, 10)
```

```python
get_nmf_topics(model3, 10)
```

## Что по NMF...

Лучше всего справилась третья модель **NMF(n_components=60, init='nndsvdar', solver='mu')**

Давайте выведем топ3 топика (хотя тут они все очень хорошие).

```python
for el in ['Topic # 21', 'Topic # 23', 'Topic # 09']:
    print('{}:\n\n{}\n_______\n'.format(el, '\n'.join(get_nmf_topics(model3, 10)[el].values)))
```
